```python
import simple_message as simple_message
import multiline_message as multiline_message

def run_block_a():
    print("Which program in  'Week 1: Basics - Output' do you wish to run?")
    print("---------------------------------------------------------------")
    print("simple_message")
    print("multiline_message")
    
    response = input()
    if response == 'simple_message':
        simple_message.run()
    elif response == 'multiline_message':
        multiline_message.run()
        
def run():
    
    while(True):
        print("What would you like to do?")
        print("[a] Run 'Week 1: Basics programs - output'")
        print("[q] quit")
        response = input()
        
        if response == 'a':
            run_block_a()
        elif response == 'q':
            break
        else:
            print("Invalid option, please try again!")
            
if __name__ == "__main__":
    run()
```
